```python
import pandas as pd
import panel as pn
import hvplot.pandas

# Carga de datos
pool_1 = pd.read_csv('C:\\Users\\andre\\Documents\\Sensor Pandas\\153560.csv', delimiter=',')
pool_2 = pd.read_csv('C:\\Users\\andre\\Documents\\Sensor Pandas\\154455.csv', delimiter=',')
pool_3 = pd.read_csv('C:\\Users\\andre\\Documents\\Sensor Pandas\\154569.csv', delimiter=',')
pool_4 = pd.read_csv('C:\\Users\\andre\\Documents\\Sensor Pandas\\153560.csv', delimiter=',')
pool_5 = pd.read_csv('C:\\Users\\andre\\Documents\\Sensor Pandas\\154455.csv', delimiter=',')
pool_6 = pd.read_csv('C:\\Users\\andre\\Documents\\Sensor Pandas\\154569.csv', delimiter=',')

# Funcin para crear el grfico y devolver el nombre descriptivo
def create_plot(pool, eje_y):
    data = None
    if pool == 'Pool 1':
        data = pool_1
    elif pool == 'Pool 2':
        data = pool_2
    elif pool == 'Pool 3':
        data = pool_3
    elif pool == 'Pool 4':
        data = pool_4
    elif pool == 'Pool 5':
        data = pool_5
    elif pool == 'Pool 6':
        data = pool_6

    if data is not None:
        data['time_stamp'] = pd.to_datetime(data['time_stamp'])
        data['time_stamp'] = data['time_stamp'].dt.round('D')

        columna_elegida = eje_y
        plot = data.hvplot(
            x='time_stamp',
            xlabel='Marca Temporal',
            y=columna_elegida,
            title=f'Indicador vs {columna_elegida}')
        
        return plot, f'{pool} - {columna_elegida}'  # Devuelve el grfico y el nombre
    else:
        return None, None

# Crear los widgets para seleccionar el pool y el eje Y
pool_selector = pn.widgets.RadioButtonGroup(
    name='Seleccione el Pool',
    options=['Pool 1', 'Pool 2', 'Pool 3', 'Pool 4', 'Pool 5', 'Pool 6'],
    button_type='success'
)

eje_y = pn.widgets.RadioButtonGroup(
    name='eje Y',
    options=['humidity', 'temperature', 'pm2.5_atm', 'pm10.0_atm'],
    button_type='success'
)

# Funcin para actualizar el grfico cuando cambian las selecciones
def update_plot(change):
    pool = pool_selector.value
    eje_y_value = eje_y.value
    plot, plot_name = create_plot(pool, eje_y_value)
    
    if plot:
        dashboard[1] = plot
        dashboard[0] = pn.Row(pool_selector, eje_y)  # Actualiza el ttulo con el nombre descriptivo
        dashboard[2] = pn.pane.Markdown(f'Grfico seleccionado: {plot_name}')

pool_selector.param.watch(update_plot, 'value')
eje_y.param.watch(update_plot, 'value')

# Valores iniciales
initial_pool = 'Pool 1'
initial_eje_y = 'humidity'
initial_plot, initial_plot_name = create_plot(initial_pool, initial_eje_y)

# Crear el panel de control y el panel del grfico inicial
dashboard = pn.Column(
    pn.Row(pool_selector, eje_y),
    initial_plot,
    pn.pane.Markdown(f'Grfico seleccionado: {initial_plot_name}')
)

dashboard.servable()
```
